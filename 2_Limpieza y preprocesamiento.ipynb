## 1.- Librerías

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

## 2.- Cargar archivo de datos

```python
raw_data = pd.read_csv('Data/Raw Data.csv', sep=";", header='infer', decimal=",", na_values='?')        
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data.describe()
```

## 3.- Tratamiento de datos vacíos

```python
total = raw_data.count().sum()
print('Número de valores en el DataFrame: ',total)
vacios = raw_data.isna().sum().sum()
print('Número de NaN en el DataFrame: ',vacios)
porcentaje= round((1-(abs((vacios-total))/total))*100,2)
print('Datos faltantes en el DataFrame: ',porcentaje, "%")
```

```python
sns.heatmap(raw_data.isna(), cbar=False)
```

```python
# Remplazo valores vacíos mediante una interpolación lineal
# https://pandas.pydata.org/pandas-docs/stable/user_guide/missing_data.html
data = raw_data
data = data.interpolate()

print("Número de NaN en el DataFrame: " + str(data.isna().sum().sum()))
```

## 4.- Conversión a estampa de tiempo

```python
# Conversión de tiempo a marca de tiempo (un objeto de fecha y hora)
data['Timestamp'] = pd.to_datetime(data['Time'],format='%d/%m/%Y %H:%M')

# Asegurarse de que no haya lagunas temporales
plt.figure(num=1, figsize=(10,5))
data.Timestamp.plot() #Trazado de la columna de marca de tiempo recién agregada
plt.xlabel('Recuento de lectura')
plt.ylabel('Fecha')
plt.show()
```

```python
# Obtener la diferencia mínima entre dos marcas de tiempo consecutivas:
min = str(np.min(np.diff(data['Timestamp'])))
min_int = str(float(min.split()[0])/60000000000)
print("La diferencia mínima entre dos marcas de tiempo consecutivas: " + min + " - " + min_int +" minutos")

# Obtener la diferencia máxima entre dos marcas de tiempo consecutivas:
max = str(np.max(np.diff(data['Timestamp'])))
max_int = str(float(max.split()[0])/60000000000)
print("La diferencia máxima entre dos marcas de tiempo consecutivas: " + max  + " - "+ max_int +" minutos")
```

```python
data = data.drop(['Time'],axis=1)
```
